### Task 3: Data Visualization


```python
#Create barchart
plt.figure(figsize=(10,6))

ax = sns.barplot(
    x= df_manuc.values,
    y=df_manuc.index,
    orient="h",
    palette = "viridis"
)
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10)

plt.ylabel("Manufacturer")
plt.xlabel("Number of cars")
plt.title("Manufacturer Per Cars");
```

```python
plt.hist(df['Price'], bins=10)
plt.title("Price Distribution")
plt.xlabel('price');
```

```python
#plot scatter plot
plt.scatter(
    x=df['Mileage'],
    y= df['Price']
)
#plot title
plt.title("Milleage VS Price")
#plot x axis label
plt.xlabel('Mileage')

# y-axis label
plt.ylabel('Price [USD]');
```

```python
#lineplot
sns.lineplot(x=df['Year of manufacture'],y=df['Price'])
```
